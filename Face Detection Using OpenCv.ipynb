# Face Detection Using OpenCv

```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Using cascade classifier and loads nose xml file to identify nose points
nose_cascade = cv2.CascadeClassifier('Cascade Classifier/Nose18x15.xml')
```

```python
# Using cascade classifier and loads eyes xml file to identify eyes points
eyes_cascade = cv2.CascadeClassifier('Cascade Classifier/frontalEyes35x16.xml')
```

```python
face_cascade = cv2.CascadeClassifier('Cascade Classifier/haarcascade_frontalface_default.xml')
```

```python
# input image
img = plt.imread('Cascade Classifier/shubham_before.jpg')
img1 = img.copy()
```

<h3>Image Before detection</h3>

```python
# display image
plt.imshow(img)
```

```python
img.shape
```

```python
# Co-ordinates where eyes are detected
eye = eyes_cascade.detectMultiScale(img)[0]
print(eye)
```

```python
eye_x, eye_y, eye_w, eye_h = eye
```

```python
# Co-ordinates where nose is detected
nose = nose_cascade.detectMultiScale(img, 1.5)[0]
print(nose)
```

```python
nose_x, nose_y, nose_w, nose_h = nose
```

```python
# Co-ordinates where mouth is detected
face = face_cascade.detectMultiScale(img)[0]
print(face)
```

```python
face_x, face_y, face_w, face_h = face
```

<h3>Drawing the rectanle on detected face part</h3

```python
# detecting & drawing the rectangle on eyes portion
img_detect = cv2.rectangle(img, (eye_x, eye_y), (eye_x + eye_w, eye_y + eye_h), (255, 255, 0), 3)
```

```python
# detecting & drawing the rectangle on nose portion
img_detect = cv2.rectangle(img, (nose_x, nose_y), (nose_x + nose_w, nose_y + nose_h), (255, 255, 0), 3)
```

```python
# detecting & drawing the rectangle on mouth portion
img_detect = cv2.rectangle(img, (face_x, face_y), (face_x + face_w, face_y + face_h), (255, 0, 0), 3)
```

```python
plt.imshow(img_detect)
```
